## Calculation of the Low Resistance

The circuit for the calculation of the low resistance is shown in the figure below.
***<p style="text-align: center;"></p>***

Here r is the small resistance value we want to measure, and R is a known resistor whose value is much greater than r (r ≪ R). The current through the circuit for a given applied voltage V is given by:
\begin{equation}
I = \frac{V}{R + r}
\end{equation}

Since r ≪ R we can approximate R + r = R. Now the current becomes: $I=V/R$. As the resistances are connected in series, the current through them will be the same. Therefore,

\begin{equation}
\frac{V_R}{R} = \frac{V_r}{r} = \frac{V}{R}
\end{equation}

The output of the lock in amplifier is $V_{out} = \alpha V_r$. So the relation between the $V_{out}$ and $V_r$ is given by:
\begin{equation}
V_{out} =  \frac{\alpha r}{R} V
\end{equation}

Where alpha is obtained from te calibration

```python
freq_list=[200,500,1000]
color_list=['r','g','b']
########################################################################## 
data_dict_cal = {}
for freq in freq_list:
    x_points = f'xlist_{freq}'
    y_points = f'ylist_{freq}'
    fit_stuff= f'fit_{freq}'
    err_stuff= f'err_{freq}'
    data_dict_cal[x_points] = []
    data_dict_cal[y_points] = []
    data_dict_cal[fit_stuff] = []
    data_dict_cal[err_stuff] = []
########################################################################## 
import matplotlib.pyplot as plt_cal
fig_cal,cal = plt_cal.subplots(figsize=(16,9))
i=0
for f in freq_list:
    data=np.loadtxt(f"Data\\LR_dat_{f}Hz.txt",usecols=(0,1))
    cal.plot(data[:,0],data[:,1],label=f"{f}Hz",color=color_list[i])
    cal.scatter(data[:,0],data[:,1],s=10,color=color_list[i])
    data_dict_cal[f'xlist_{f}'] = data[:,0]
    data_dict_cal[f'ylist_{f}'] = data[:,1]
    i+=1
##########################################################################     
plt_cal.xlabel(r"Input voltage ($V_{in}$)(RMS)", fontsize=16)
plt_cal.ylabel(r"Output voltage ($V_{out}$)(RMS)", fontsize=16)
plt_cal.title(r"Plot of $V_{in}$ vs $V_{out}$ for Low Resistance",fontweight='bold', fontsize=18)
plt_cal.legend()
plt_cal.grid()
plt_cal.show()
##########################################################################
```
