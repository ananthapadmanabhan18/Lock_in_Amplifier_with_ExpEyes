# __List of Functions for the Expeyes__

#### Importing and opening the expeyes

```python
import eyes17.eyes

p=eyes17.eyes.open()
```

### For Setting output from Expeyes

Sets the DC voltages at PV1 and PV2. PV1 range is -5 to 5. PV2 range is -3.3 to 3.3

```python
v=p.set_pv1(4) # set voltage to 4V
v=p.set_pv2(4) # set voltage to 4V
```

### For Acquiring Voltages

```python
v= p.get_voltage('A1')  # get voltage from A1
v= p.get_voltage('A2')  # get voltage from A2
v= p.get_voltage('A3')  # get voltage from A3
v= p.get_voltage('MIC') # get voltage from MIC
v= p.get_voltage('SEN') # get voltage from SEN
```

### Acquiring Frquencies

```python
p.get_freq('IN2') # Measures the frequency of a wave on the input, IN2 or SEN.
```

### For Acquiring Other Stuffs

```python
t=p.get_voltage_time('A1') #Returns a tuple, containing the computerâ€™s time stamp and the voltage at the specified input
R=p.get_resistance('A1')   #Returns the value of resistance connected to SEN, 
                           #it should be between 100 Ohm and 100k for reasonable accuracy.
C=p.get_capacitance('A1')  #Returns the value of capacitance connected to IN1 (works well in pF ranges)
```

### Setting the Output Voltages

```python
p.set_state(OD1=1) #Sets the output of OD1, SQ1 etc. to 1 (high)
p.set_sine_amp(2)  #The amplitude can be set to 3 pre-defined values of the peak voltage ( 0-> 80mV, 1-> 1V, 2-> 3V) 
```

### Setting Output Frequencies

```python
p.set_sine(502)      # Generates the sinewave of requested frequency on WG (range from 5Hz to 5000Hz).
p.set_sqr1_slow(0.5) # Sets the frequency of SQ1 output (range from 0.1Hz to 1 MHz). Resolution 
                     # is high but WG is disabled when SQ1 is operated in this mode.
p.set_sqr2_slow(0.5) #Similar to set_sqr1() but SQ2 is not available along with WG, only one at a time.
p.set_sqr1(1000, 30) #Sets the frequency of SQ1 output (range from 0.1Hz to 1 MHz). here f=1000Hz, duty=30%
```
