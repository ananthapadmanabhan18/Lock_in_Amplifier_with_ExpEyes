```python
import numpy as np
import scipy as sp
```

```python
data_mono = np.loadtxt('data_mono.txt')
data_poly = np.loadtxt('data_poly.txt')

def f(x,a,b):
    return a*x+b
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,5))
plt.scatter(data_poly[:,0], data_poly[:,6], label='Polycrystaline',s=15,color='black')
plt.plot(data_poly[:,0], data_poly[:,6], color='red')
plt.xlabel(r'$V_R$ (V)')
plt.ylabel(r'C (nF)')
plt.title('Capacitance vs Reverse Voltage for polycrystaline Solar cell', fontsize=12)
plt.legend()
plt.grid()
plt.show()





import matplotlib.pyplot as plt
plt.figure(figsize=(9,5))
plt.scatter(data_poly[:,0], data_poly[:,7], label='Polycrystaline',s=15,color='black')
x=np.linspace(0.1,2.1,100)
y=f(x,53.212,-17.336)
plt.plot(x,y, color='red',label='Linear Fit')
plt.xlabel(r'$V_R$ (V)')
plt.ylabel(r'$1/C^2$ (nF)')
plt.title(r'$1/C^2$ vs $V_R$ for polycrystaline Solar cell', fontsize=12)
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,5))
plt.scatter(data_mono[:,0], data_mono[:,6], label='Polycrystaline',s=15,color='black')
plt.plot(data_mono[:,0], data_mono[:,6], color='red')
plt.xlabel(r'$V_R$ (V)')
plt.ylabel(r'C (nF)')
plt.title('Capacitance vs Reverse Voltage for Monocrystalline Solar cell', fontsize=12)
plt.legend()
plt.grid()
plt.show()





import matplotlib.pyplot as plt
plt.figure(figsize=(9,5))
plt.scatter(data_mono[:,0], data_mono[:,7], label='Monocrystalline',s=15,color='black')
x=np.linspace(0.1,2.1,100)
y=f(x,5.743692249,-2.880222429)
plt.plot(x,y, color='red',label='Linear Fit')
plt.xlabel(r'$V_R$ (V)')
plt.ylabel(r'$1/C^2$ (nF)')
plt.title(r'$1/C^2$ vs $V_R$ for Monocrystalline Solar cell', fontsize=12)
plt.legend()
plt.grid()
plt.show()
```
